Data Cleaning

```python
df = df.dropna()
df.shape
```

```python
df['rating'].value_counts()
```

Changing format of Date

```python
# Convert all values in the 'date_added' column to strings
df['date_added'] = df['date_added'].astype(str)

# Strip leading and trailing whitespace from the 'date_added' column
df['date_added'] = df['date_added'].str.strip()

# Convert to datetime
df["date_added"] = pd.to_datetime(df['date_added'], errors='coerce')

# Extract day, month, and year
df['day_added'] = df['date_added'].dt.day
df['year_added'] = df['date_added'].dt.year
df['month_added'] = df['date_added'].dt.month

# Fill NaT values in 'year_added' with a placeholder (e.g., 0)
df['year_added'].fillna(0, inplace=True)

# Ensure the year_added column is of integer type
df['year_added'] = df['year_added'].astype(int)

```

Different types

Type: Movie and TV Shows

```python
custom_palette = ['#FF6347', '#4682B4', '#32CD32', '#FFD700', '#EE82EE']
sns.countplot(x='type', data=netflix, palette=custom_palette, orient='v')
fig = plt.gcf()
fig.set_size_inches(10,10)
plt.title('Type')
```

Rating of shows and movies

```python
custom_palette = ['#FF6347', '#4682B4', '#32CD32', '#FFD700', '#EE82EE', '#FFA500', '#8A2BE2', '#00CED1', '#FF1493', '#7FFF00']
sns.countplot(x='rating', data=netflix, palette=custom_palette[:len(netflix['rating'].unique())])
plt.xticks(rotation=90, ha="right")
fig = plt.gcf()
fig.set_size_inches(13, 13)
plt.title('Rating')
plt.show()
```

Relation between Type and Rating

```python
plt.figure(figsize=(10,8))
sns.countplot(x='rating',hue='type',data=netflix)
plt.title('Relation between Type and Rating')
plt.show()
```

Pie-chart for the Type: Movie and TV Shows

```python
labels = ['Movie', 'TV show']
size = netflix['type'].value_counts()
colors =  ['blue', 'red']
explode = [0, 0.1]
plt.rcParams['figure.figsize'] = (9, 9)
plt.pie(size,labels=labels, colors = colors, explode = explode, shadow = True, startangle = 90)
plt.title('Distribution of Type', fontsize = 25)
plt.legend()
plt.show()
```

Pie-chart for Rating

WordCloud

```python
from wordcloud import WordCloud
```

```python
df['country'] = df['country'].astype(str).fillna('')

# Generate the word cloud
wordcloud = WordCloud(
    background_color='white',
    width=1920,
    height=1080
).generate(" ".join(df['country']))

# Plot the word cloud
plt.subplots(figsize=(25, 15))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
```

Top 15 Countries contributing on Netflix

```python
top_countries = netflix['country'].value_counts().head(15)
top_countries = top_countries.sort_values(ascending=False)
plt.figure(figsize=(10, 8))
top_countries.plot(kind='barh', color='red')
plt.xlabel('Number of Titles')
plt.ylabel('Country')
plt.title('Top 15 Countries in Netflix')
plt.tight_layout()
plt.show()
```

Top 10 Actors

```python
movies_df = netflix[netflix['type'] == 'Movie']
actors = movies_df['cast'].str.split(', ').explode()
top_actors = actors.value_counts().head(10)
plt.figure(figsize=(10, 6))
top_actors.plot(kind='bar', color='blue')
plt.xlabel('Actor')
plt.ylabel('Number of Appearances')
plt.title('Top 10 Movie Actors on Netflix')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

Top 10 Directors

```python
movies_df = netflix[netflix['type'] == 'Movie']
directors = movies_df['director'].str.split(', ').explode()
top_directors = directors.value_counts().head(10)
print("Top 10 Directors on Netflix:")
print(top_directors)
```

```python
netflix['date_added'] = pd.to_datetime(netflix['date_added'], errors='coerce')
netflix['year_added'] = netflix['date_added'].dt.year
titles_added_movies = netflix[netflix['type'] == 'Movie']['year_added'].value_counts().sort_index()
titles_added_tv_shows = netflix[netflix['type'] == 'TV Show']['year_added'].value_counts().sort_index()
titles_added_total = netflix['year_added'].value_counts().sort_index()
plt.figure(figsize=(12, 8))
plt.plot(titles_added_movies.index, titles_added_movies.values, marker='o', linestyle='-', color='blue', label='Movies')
plt.plot(titles_added_tv_shows.index, titles_added_tv_shows.values, marker='o', linestyle='-', color='green', label='TV Shows')
plt.plot(titles_added_total.index, titles_added_total.values, marker='o', linestyle='-', color='purple', label='Total')
plt.title('Number of Titles Added on Netflix Over the Years')
plt.xlabel('Year')
plt.ylabel('Number of Titles Added')
plt.grid(True)
plt.xticks(titles_added_total.index)
plt.legend()
plt.tight_layout()
plt.show()
```
